```python
%sql
create table git_catalog.git_schema.git_table (
  start_location string,
  end_location string,
  distance string
) using delta
```

```python
%sql
insert into git_catalog.git_schema.git_table values('chennai','mumbai','500'),('mumbai','chennai','500'),('patna','ranchi','300'),('bangalore','chennai','350'),('chennai','bangalore','350')
```

```python
%sql
select * from git_catalog.git_schema.git_table
```

```python
%sql
select 
  case when start_location < end_location then end_location else start_location end as end_location,
  case when end_location >start_location then start_location else end_location end as start_location,distance
from git_catalog.git_schema.git_table
```
